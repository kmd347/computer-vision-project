```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/jldbc/coffee-quality-database/master/data/arabica_data_cleaned.csv"
df = pd.read_csv(url)

df.head()
```

```python
print("Dataset shape:", df.shape)
print("\nColumns:\n", df.columns)
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Total.Cup.Points"], bins=20, kde=True)
plt.title("Distribution of Coffee Quality Scores")
plt.show()
```

```python
#binary target creation 
df["high_quality"] = (df["Total.Cup.Points"] >= 85).astype(int)

df["high_quality"].value_counts()
```

```python
#class balance visualizaiton 
sns.countplot(x="high_quality", data=df)
plt.title("Class Distribution: High Quality vs Not High Quality")
plt.show()
```

```python
#correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```
